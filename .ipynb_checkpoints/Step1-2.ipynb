## 1D linear convection
\begin{align}
\frac{\partial u} {\partial t} + c\frac{\partial u} {\partial x} = 0
\end{align}

### constant transport with velocity c
initail profile : $u(x, t=0) = u_{0}(x)$  
solution : $u(x, t) = u_{0}(x - ct)$

After time t: $u(x)$ is just the initila profile displaced with distnace $x = ct$  
(=> wave propagation) : pure convection

$$\frac{\partial u}{\partial x}\approx \frac{u(x+\Delta x)-u(x)}{\Delta x}$$

### index
space : i with Backward Difference  
time : n with Forward Difference

Forward differnece : $\Delta u = u_{i+1} - u_{i}$  
Backward difference :$\Delta u = u_{i} - u_{i-1}$

### taylor expansion
expressing a function into a sum of power functions  
$$ f(x) = p_{\infty}(x) $$  
\begin{eqnarray*} p(x) & = & \sum^{\infty} _{n=0} \frac{f^{(n)} (a)} {n!}(x-a)^{(n)} \\ & = & f(a) + f'(a)(x-a) + \frac {1} {2!} f''(a)(x-a)^2 + \cdots + \frac {1} {n!}f^{(n)}(a)(x-a)^{n}
\end{eqnarray*}

Forward : $\Delta u = u_{i+1} - u_{i}$ ($p(x) = u_{i+1}, f(a) = u_{i}$)  
Backward :$\Delta u = u_{i} - u_{i-1}$ ($p(x) = u_{i-1}, f(a) = u_{i}$)

### Discrete equation
$$\frac{u^{n+1}_{i} - u^{n}_{i}} {\Delta t} + c\frac {u^{n}_{i} - u^{n}_{i-1}} {\Delta x}= 0$$

### Rearranged
$$u^{n+1}_{i} = u^{n}_{i} - c\frac {\Delta t} {\Delta x}(u^{n}_{i} - u^{n}_{i-1})$$

```python
import numpy as np
import matplotlib.pyplot as plt
import time, sys
```

```python
%matplotlib inline
```

```python
nx = 41
dx = 2 / (nx - 1)
nt = 20
dt = .025
c = 1.5
```

```python
u = np.ones(nx)
u[int(.5/dx) : int(1 / dx + 1)] = 2
print(u)
```

```python
plt.plot(np.linspace(0, 2, nx),u)
```

```python
u = np.ones(nx)
u[int(.5/dx) : int(1 / dx + 1)] = 2
un = np.ones(nx)

fig, axes = plt.subplots()

for n in range(nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - c * dt / dx * (un[i] - un[i-1])
    if not n%5:
        plt.plot(np.linspace(0,2,nx), u)
```

```python
nx = 61
dx = 2 / (nx - 1)
nt = 20
dt = .025
c = 1

u = np.ones(nx)
u[int(.5/dx) : int(1 / dx + 1)] = 2
un = np.ones(nx)

fig, axes = plt.subplots()

for n in range(nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - c * dt / dx * (un[i] - un[i-1])
    if not n%5:
        plt.plot(np.linspace(0,2,nx), u)
```

## 1D Non-linear convection
\begin{align}
\frac{\partial u} {\partial t} + u\frac{\partial u} {\partial x} = 0
\end{align}

### transport with velocity u(inviscid Burgers)
Can generate discontinuous solutions from smoooth Initial Condition  
(Similar to shock creation)

### Discrete equation
$$\frac{u^{n+1}_{i} - u^{n}_{i}} {\Delta t} + u^{n}_{i}\frac {u^{n}_{i} - u^{n}_{i-1}} {\Delta x}= 0$$

### Rearranged
$$u^{n+1}_{i} = u^{n}_{i} - u^{n}_{i}\frac {\Delta t} {\Delta x}(u^{n}_{i} - u^{n}_{i-1})$$

```python
nx = 41
dx = 2 / (nx - 1)
nt = 20
dt = 0.025

u = np.ones(nx)
u[int(0.5/dx) : int(1 / dx + 1)] = 2

un = np.ones(nx)
for n in range(nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i-1])

plt.plot(np.linspace(0, 2, nx), u)
```

```python
nx = 21
dx = 2 / (nx - 1)
nt = 20
dt = 0.025

u = np.ones(nx)
u[int(0.5/dx) : int(1 / dx + 1)] = 2

un = np.ones(nx)
for n in range(nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i-1])

plt.plot(np.linspace(0, 2, nx), u)
```
